```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def hmap(arr,cmap='Greens'):
    plt.tight_layout()
    out = sns.heatmap(arr, cmap=cmap, cbar=False, square=True, xticklabels=False, yticklabels=False, linecolor='white', linewidths=1)
    plt.show()
    return out
```

```python
arr = np.random.randn(4, 10)
cmap = 'Greens'
img = hmap(arr, cmap)
img.figure.savefig(f"../assets/heatmap-{cmap}.png")
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate a 2D grid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)

# Compute the 2D normal distribution
z = (1 / (2 * np.pi)) * np.exp(-0.5 * (x**2 + y**2))

# Plot the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='Greys', edgecolor='none')

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
# ax.grid(False)
ax.set_xlabel('z1')
ax.set_ylabel('z2')
ax.set_zlabel('z3')

fig.savefig("../assets/3d_surface_plot.png")
plt.show()
```

```python
from sklearn.decomposition import PCA

# Generate a 2D blob of data with unequal variances
mean = [0, 0]
cov = [[5, 3], 
       [3, 4]]  # Covariance matrix with unequal variances
data = np.random.multivariate_normal(mean, cov, size=500)

# Perform PCA
pca = PCA(n_components=2)
pca.fit(data)
components = pca.components_
explained_variance = pca.explained_variance_

# Plot the data and principal components
plt.figure(figsize=(3, 3))
plt.scatter(data[:, 0], data[:, 1], alpha=0.1, color='ForestGreen', label='Data points')

# Plot the principal components as arrows
for i, (component, variance) in enumerate(zip(components, explained_variance)):
    plt.arrow(0, 0, component[0] * variance, component[1] * variance, color='black',head_width=0.5, label=f'PC{i+1}')

# plt.legend()
plt.gca().set_frame_on(False)
plt.xticks([])
plt.yticks([])
plt.savefig("../assets/pca-example.png")
plt.show()
```
